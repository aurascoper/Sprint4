# Exploratory Data Analysis of US Vehicles Dataset

```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('../vehicles_us.csv')
df.head()
```

```python
df.info()
```

```python
px.histogram(df, x='price', nbins=50, title='Price Distribution')
```

```python
px.histogram(df, x='model_year', nbins=30, title='Model Year Distribution')
```

```python
px.scatter(df, x='odometer', y='price', title='Price vs Odometer')
```

```python
px.scatter(df, x='model_year', y='price', color='fuel', title='Price vs Model Year by Fuel Type')
```
